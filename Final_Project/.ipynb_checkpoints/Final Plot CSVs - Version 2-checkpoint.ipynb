```python
import numpy as np
import librosa
import aubio
import matplotlib.pyplot as plt
import librosa.display
import pandas
import os
import glob
```

```python
def extractAttr(filename):
    attr = pandas.read_csv(filename,header=None)
    x = attr.iloc[:,0]
    y = attr.iloc[:,1]
    yz = np.nonzero(y)
    #print(yz)
    xnew = []
    ynew = []
    for item in yz:
        xnew.append(x[item])
        ynew.append(y[item])
    mean = np.mean(ynew)
    sd = np.std(ynew)
    return mean, sd
```

```python
def readall(path):
    meanarray = []   
    sdarray = []
    namearray = []
    meanofmeans = []
    maximum = 0
    minimum = 1000
    for subdir, dirs, files in os.walk(path):
        array1 =[]
        array2 =[]
        
        namearray.append(subdir)
        for filename in files:
            if filename.endswith(".csv"):    
                #print(subdir)
                #print(filename)
                filepath = filepath = subdir + os.sep + filename
                mean,sd = extractAttr(filepath)
                array1.append(mean)
                array2.append(sd)
                if mean > maximum:
                    maximum = mean
                    #print(subdir+"/"+filename)
                    maxfile = subdir+filename
                if mean < minimum:
                    minimum = mean
                    minfile = subdir+"/"+filename
        meanarray.append(array1)
        sdarray.append(array2)
        meanofmeans.append(np.mean(array1))
        
    
    return meanarray,sdarray,namearray, maximum, maxfile, minimum, minfile, meanofmeans
```

```python
def plotMeans(name):
    path = '/Users/mattweisberg/GitHub/CompMusFall2018/Final_Project/All/' + name
    #path = os.getcwd()
    means, sds, namearray, maximum, maxfile, minimum, minfile,meanofmeans = readall(path)
    plt.figure()
    #plt.subplot(2,1,1)
    plt.plot(means[1], '*r', label=namearray[1])
    plt.plot(means[2], '*g',label=namearray[2])
    plt.plot(means[3], '*b',label=namearray[3])
    plt.plot(means[4], '*k',label=namearray[4])
    plt.plot(means[5], '*y',label=namearray[5])
    plt.title(name)
    plt.legend(loc="best")
    plt.legend("AADGM")
    
    plt.tight_layout()
    plt.savefig(name + '.png')
    #plt.subplot(2,1,2)
    plt.figure()
    plt.title("Mean of Means" +" " + name)
    plt.plot(meanofmeans[1], '*r', label=namearray[1])
    plt.plot(meanofmeans[2], '*g',label=namearray[2])
    plt.plot(meanofmeans[3], '*b',label=namearray[3])
    plt.plot(meanofmeans[4], '*k',label=namearray[4])
    plt.plot(meanofmeans[5], '*y',label=namearray[5])
    plt.legend(loc="best")
    plt.legend("AADGM")
    
    plt.savefig("Mean of Means" + name + '.png')
    plt.tight_layout()
    
    print("/")
    print(maximum, maxfile)
    print("/")
    print(minimum, minfile)
    
```
