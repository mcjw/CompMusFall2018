```python
import numpy as np
import librosa
import librosa.display
import matplotlib.pyplot as plt

file = 'avm.wav'
sig , sr = librosa.load(file,mono=True,sr=None)
```

```python
print(sr)
```

```python
avmSpec = librosa.amplitude_to_db(np.abs(librosa.stft(sig,n_fft=2048, hop_length=32, win_length=1024)), ref=np.max)
librosa.display.specshow(avmSpec, x_axis = 'time' , y_axis='log')
plt.colorbar(format='%+02.0f dB')
plt.title('AVM Log-frequency Power spectrogram')
```

```python
s = librosa.feature.melspectrogram(y=sig, sr=sr, n_fft=2048, n_mels=128)
log_s = librosa.power_to_db(s, ref=np.max)
librosa.display.specshow(log_s, sr=sr, x_axis='time', y_axis='mel')
plt.title('AVM Mel Power spectrogram')
plt.colorbar(format='%+02.0f dB')
```

```python
mfcc = librosa.feature.mfcc(y=sig, sr=sr, S=log_s, n_mfcc=13)
librosa.display.specshow(mfcc[1:12,:],x_axis='time')
plt.ylabel('MFCCs (2:13)')
plt.colorbar()
```

```python
stftChroma = librosa.feature.chroma_stft(y=sig,sr=sr)
librosa.display.specshow(stftChroma, y_axis='chroma', sr=sr,vmin=0.0, vmax=1.0, x_axis='time')
plt.colorbar()
```
